## Importing Libraries

```python
## Loading and preprocessing
import numpy as np
import pandas as pd

## For Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_new = pd.read_csv('new_cleaned_toyota_1.csv')
```

```python
df_new = df_new.drop('Unnamed: 0.1', axis=1)
```

```python
df_new.to_csv('new_cleaned_toyota_1.csv')
```

```python
df_new['FuelType'] = df_new['FuelType'].astype('category')
df_new['MetColor'] = df_new['MetColor'].astype('category')
df_new['Automatic'] = df_new['Automatic'].astype('category')
df_new['Doors'] = df_new['Doors'].astype('category')
df_new['CC'] = df_new['CC'].astype('category')

df_new.dtypes
```

## Target Variable Analysis

## Frequency Table Using Histogram

```python
## Using matplotlib to create a frequency distribution table
plt.figure(figsize=(8,4))
plt.title("Frequency Distribution of Price of Car")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.hist(df_new['Price'],color='yellow',bins=10)
plt.show()
```

```python
## Using Normalised to create a frequency table
new_price = np.log1p(df_new['Price'])
```

```python
plt.figure(figsize=(8,4))
plt.title("Frequency Distribution of Normalised Prices of Car")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.hist(df_new['Price'],color='orange',bins=10)
plt.show()
```

## Plotting Using Seaborn

```python
## Using seaborn to create frequency table on histogram
plt.figure(figsize=(8,4))
sns.histplot(df_new['Price'],color='purple',bins=30)

plt.title("Frequency Distribution of Price of Car")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

```python
## Using Seaborn to create scatter plot
plt.figure(figsize=(8,4))
sns.scatterplot(data=df_new, x='Age', y='Price',color='green')

plt.title("Correlation b/n Age and Price")
plt.xlabel("Age(months)")
plt.ylabel("Price")

plt.show()
```

```python
## Using Seaborn to create scatter plot
plt.figure(figsize=(8,4))
sns.boxplot(x='FuelType',y='Price', data=df_new, hue='Automatic')

plt.title("Distribution of Fuel Type")
plt.xlabel("Fuel Type")
#plt.ylabel("Price")

plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='FuelType', data=df_new, hue='Automatic')

plt.title("Distribution of Fuel Type")
plt.xlabel("Fuel Type")
#plt.ylabel("Price")

plt.show()
```
